# 4番目に実行



```
class LR2LampGraph:
    def __init__(self):
        self.__get_account_info()

    def __get_account_info(self):
        self.lr2id = str(LR2ID)
        self.passmd5 = hashlib.md5(PASSWORD.encode('utf-8')).hexdigest()

        url = f"http://www.dream-pro.info/~lavalse/LR2IR/restorescore.cgi?id={self.lr2id}&passmd5={self.passmd5}"

        req = urllib.request.Request(url)

        with urllib.request.urlopen(req) as res:
            body = res.read().decode("utf-8")

        root = ET.fromstring(body)

        self.lamp_dict = {}

        for score in root.iter("score"):
            self.lamp_dict[score.find("hash").text] = int(
                score.find("clear").text)

    def get_lamp_graph(self, url):
        print(url)
        table = lr2irscraper.get_bms_table(url)
        table_dict = table.to_dict()
        print(table_dict["header"]["symbol"], table_dict["header"]["name"])

        # F H C E F N
        lamps = {}

        for bms in table_dict["data"]:
            if not bms["level"] in lamps:
                lamps[bms["level"]] = [0, 0, 0, 0, 0, 0]
            lamps[bms["level"]][5 - self.lamp_dict.get(bms["md5"], 0)] += 1

        labels = []
        percents = [[] for i in range(6)]
        for level, lamp in lamps.items():
            if sum(lamp) > 0:
                total_lamps = sum(lamp)
                labels.append(str(level))
                for i in range(6):
                    percents[i].append(lamp[i] * 100 / total_lamps)

        plt.clf()
        plt.figure(figsize=(7,7))

        y = np.arange(len(labels))
        width = 0.7

        p = []
        cumsum = [0 for i in range(len(labels))]
        for i in range(6):
            p.append(plt.barh(y, percents[i], width, left=cumsum, color=(
                "#CCFFFF", "#FFCCCC", "#88CCFF", "#CCFFCC", "#CCCCCC", "#FFFFFF")[i]))
            for j in range(len(labels)):
                cumsum[j] += percents[i][j]

        plt.xlabel("%")
        plt.ylabel(table_dict["header"]["symbol"])
        plt.title(table_dict["header"]["name"])
        plt.yticks(y, labels)
        plt.xticks(np.arange(0, 101, 10))
        plt.legend((p[i][0] for i in range(6)), ("F-COMBO",
                                                 "H-CLEAR", "CLEAR", "E-CLEAR", "FAILED", "NO PLAY"))


        plt.savefig(table_dict["header"]["name"] + ".png", transparent=False)
        #plt.show()


lr2lampgraph = LR2LampGraph()
```
